```python
#คำนวณค่าไฟฟ้าการไฟฟ้าส่วนภูมิภาค
units=float(input("จำนวนหน่วยไฟฟ้าที่ใช้"))
#อัตราค่าไฟฟ้าต่อหน่วย
units1=3.2484 #1-150หน่วย
units2=4.2218 #151-400 หน่วย
units3=4.4217 #410หน่วยขึ้นไป

service_Charge=24.62 #ค่าบริการรายเดือน/บาท
Ft=0.3672 #ค่าไฟฟ้าผันแปร(แต่ละเดือน)
vat=0.07 #ค่าอัตราภาษีมูลค่าเพิ่ม 7%

def bill(units):
  #คำนวณค่าไฟฟ้าตามหน่วย
  if units<=150:
    bill=units*units1
  elif units<=400:
    bill=(150*units1)+((units-150)*units2)
  else:
    bill=(150*units1)+(250*units2)+((units-400)*units3)

  #ค่าไฟฟ้าฐาน
  Based_Amount=bill+service_Charge

  #คำนวณค่าFt
  Ft1=units*Ft

   #รวมค่าไฟฟ้า
  sub_total=Based_Amount+Ft1

   #คำนวณ vat 7%
  total_vat=sub_total*vat

  #ค่าไฟรวมทั้งหมด
  total_bill=sub_total+total_vat

  print(f"ค่าบริการรายเดือน:{service_Charge}บาท")
  print(f"ค่าไฟฟ้าฐาน{Based_Amount:.2f}บาท")
  print(f"ค่าFt{Ft1:.2f}บาท")
  print(f"ค่าไฟฟ้า:{sub_total:.2f}บาท")
  print(f"ภาษีมูลค่าเพิ่ม(vat 7%):{total_vat:.2f}บาท")
  print(f"ค่าไฟฟ้ารวมทั้งหมด:{total_bill:.2f}บาท")
  return
bill(units)

```
